# Importing the required libraries

```python
import numpy as py
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importing and loading the dataset


```python
train=pd.read_csv("https://raw.githubusercontent.com/AnirbanD24/Machine-Learning-CSE-4020/main/train.csv")
train.info()
```

```python
train.head(50)
```

# Seeing the number of null values for each field

```python
train.isnull().sum()
```

# Data Cleaning by dropping the missing values

```python
# dropping the missing values (not advisable)
df=train.dropna()
df.info()
```

```python
df.isnull().sum()
```

# Analysing the boxplot and distribution plot for Order Quantity

```python
sns.boxplot(x=train["Order Quantity"])
plt.show()
```

```python
sns.displot(x=train["Order Quantity"])
plt.show()
```

# Before removing missing values (NaN) for Order Quantity

```python
train.head(10)
```

# Finding the median for Order Quantity

```python
median_order_quantity=train['Order Quantity'].median()
print(median_order_quantity)
```

# Filling the missing values of Order Quantity using the median value

```python
train['Order Quantity']=train['Order Quantity'].fillna(median_order_quantity)
```

# After removing missing values for Order Quantity

```python
train.head(10)
```

# Analysing the boxplot and distribution plot for Sales

```python
sns.boxplot(x=train["Sales"])
plt.show()
```

```python
sns.displot(x=train["Sales"])
plt.show()
```

# Before removing missing values (NaN) for Sales

```python
train.head(10)
```

# Finding the median for Sales

```python
median_sales=train['Sales'].median()
print(median_sales)
```

# Filling the missing values of Sales using the median value

```python
train['Sales']=train['Sales'].fillna(median_sales)
```

# After removing missing values for Sales

```python
train.head(10)
```

# Analysing the boxplot and distribution plot for profit

```python
sns.boxplot(x=train["profit"])
plt.show()
```

```python
sns.displot(x=train["profit"])
plt.show()
```

# Before removing missing values (NaN) for profit

```python
train.head(10)
```

# Finding the mean for profit

```python
mean_profit=train['profit'].mean()
print(mean_profit)
```

# Filling the missing values of profit using the mean value

```python
train['profit']=train['profit'].fillna(mean_profit)
```

# After removing missing values for profit

```python
train.head(10)
```
